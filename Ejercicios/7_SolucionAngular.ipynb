# Componente angular de la solución de la ecuación de Schrödinger para el átomo de Hidrógeno

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%pylab inline
```

```python
pylab.rcParams['figure.figsize'] = (10.0, 5.0)
```

## 1. Gráficas 2D

### 1.1 Orbital $1s: n=1, l=0, m_l=0$

```python
c_1s = 1/np.sqrt(4*np.pi)

def Y_1s(theta, phi):
    return 0*theta + c_1s
```

```python
theta = np.linspace(0, 2*np.pi, 100)
phi = 0
thetaG, phiG = np.meshgrid(theta, phi)
r = Y_1s(thetaG, phiG)
r2 = Y_1s(thetaG, phiG)**2

# plots
ax1 = plt.subplot(121, projection='polar')
ax1.plot(theta.flatten(), r.flatten(), linewidth=3)
ax1.set_rmax(0.35)
ax1.set_rticks([])
ax1.set_title("$Y_{1s}$", y=1.1)
ax1.grid(True)
ax1.set_theta_zero_location("N")
# ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~
ax2 = plt.subplot(122, projection='polar')
ax2.plot(theta.flatten(), r2.flatten(), linewidth=3, color="#34A853")
ax2.set_rmax(0.35)
ax2.set_rticks([])
ax2.set_title("$Y^2_{1s}$", y=1.1)
ax2.grid(True)
ax2.set_theta_zero_location("N")

base_fsize = 18
plt.rc('axes', titlesize=base_fsize+4)
plt.show()
```

### 1.2 Orbital $P_z: n=2, l=1, m_l=0$

```python
c_2p = np.sqrt(3/(4*np.pi))

def Y_2p(theta, phi):
    return c_2p * np.cos(theta)
```

```python
theta = np.linspace(0, 2*np.pi, 100)
phi = 0
thetaG, phiG = np.meshgrid(theta, phi)
r = Y_2p(thetaG, phiG)
r2 = Y_2p(thetaG, phiG)**2

# plots
ax1 = plt.subplot(121, projection='polar')
ax1.plot(theta.flatten(), r.flatten(), linewidth=3)
ax1.set_rmax(0.6)
ax1.set_rticks([])
ax1.set_title("$Y_{P_Z}$", y=1.1)
ax1.grid(True)
ax1.set_theta_zero_location("N")
# ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~
ax2 = plt.subplot(122, projection='polar')
ax2.plot(theta.flatten(), r2.flatten(), linewidth=3, color="#34A853")
ax2.set_rmax(0.4)
ax2.set_rticks([])
ax2.set_title("$Y^2_{P_Z}$", y=1.1)
ax2.grid(True)
ax2.set_theta_zero_location("N")

base_fsize = 18
plt.rc('axes', titlesize=base_fsize+4)
plt.show()
```

### 1.3 Orbital $d_{z^2}: n=3, l=2, m_l=0$

```python
c_3d = np.sqrt(5/(16*np.pi))

def Y_3d(theta, phi):
    return c_3d * (3*(np.cos(theta)**2) -1)
```

```python
theta = np.linspace(0, 2*np.pi, 100)
phi = 0
thetaG, phiG = np.meshgrid(theta, phi)
r = Y_3d(thetaG, phiG)
r2 = Y_3d(thetaG, phiG)**2

# plots
ax1 = plt.subplot(121, projection='polar')
ax1.plot(theta.flatten(), r.flatten(), linewidth=3)
ax1.set_rmax(0.9)
ax1.set_rticks([])
ax1.set_title("$Y_{d_{z^2}}$", y=1.1)
ax1.grid(True)
ax1.set_theta_zero_location("N")
# ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~
ax2 = plt.subplot(122, projection='polar')
ax2.plot(theta.flatten(), r2.flatten(), linewidth=3, color="#34A853")
ax2.set_rmax(0.55)
ax2.set_rticks([])
ax2.set_title("$Y^2_{d_{z^2}}$", y=1.1)
ax2.grid(True)
ax2.set_theta_zero_location("N")

base_fsize = 18
plt.rc('axes', titlesize=base_fsize+4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline  
%matplotlib notebook
%pylab
```

```python
pylab.rcParams['figure.figsize'] = (10.0, 10.0)
```

### 2.1 Orbital $1s: n=1, l=0, m_l=0$

```python
c_1s = 1/np.sqrt(4*np.pi)

def Y_1s(theta, phi):
    return 0*theta + c_1s
```

```python
theta = np.linspace(0, np.pi, 100)
phi = np.linspace(0, 2*np.pi, 2*100)
thetaG, phiG = np.meshgrid(theta, phi)
r2 = Y_1s(thetaG, phiG)**2

# cartesian coords
x = r2 * np.sin(thetaG) * np.cos(phiG)
y = r2 * np.sin(thetaG) * np.sin(phiG)
z = r2 * np.cos(thetaG)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='Oranges', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
