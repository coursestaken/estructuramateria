# Gráficas de la componente radial de la solución de la ecuación de Schrödinger para el átomo de Hidrógeno

```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
pylab.rcParams['figure.figsize'] = (11.0, 5.0)
```

```python
# Constantes
a0 = 52.9 # Radio de Boht en picometros (1e-12)
a0_inv = 1/a0
```

## 1. Orbital $1s: n=1, l=0, m_l=0$

```python
# componente radial para el orbital 1s
K_1s = 2 * np.sqrt(a0_inv**3)
def R_1s(r):
    return K_1s * np.exp(-r/a0)
```

```python
r = np.linspace(0, 800, 100) # picometros

ax1 = plt.subplot(121)
ax1.plot(r, R_1s(r), linewidth=3, color="#20639F")
ax1.set_xlabel('$r$ [pm]')
ax1.set_ylabel('$R$')
ax1.grid()
# ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~
ax2 = plt.subplot(122)
ax2.plot(r, r*r*R_1s(r)**2, linewidth=3, color="#34A853")
ax2.fill_between(r, r*r*R_1s(r)**2, 0, alpha=0.2, color="#34A853")
ax2.set_xlabel('$r$ [pm]')
ax2.set_ylabel('$r^2 R^2$')
ax2.grid()

base_fsize = 12
plt.rc('axes', labelsize=base_fsize+6)
plt.rc('xtick', labelsize=base_fsize)
plt.rc('ytick', labelsize=base_fsize)
plt.tight_layout(w_pad=2)
plt.show()
```

## 2. Orbital $2s: n=2, l=0, m_l=0$

```python
# componente radial para el orbital 2s
K_2s = 1/(2*np.sqrt(2)) * np.sqrt(a0_inv**3)
def R_2s(r):
    return K_2s * (2 -r/a0) * np.exp(-r/(2*a0))
```

```python
r = np.linspace(0, 800, 100) # picometros

ax1 = plt.subplot(121)
ax1.plot(r, R_2s(r), linewidth=3, color="#20639F")
ax1.set_xlabel('$r$ [pm]')
ax1.set_ylabel('$R$')
ax1.grid()
# ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~
ax2 = plt.subplot(122)
ax2.plot(r, r*r*R_2s(r)**2, linewidth=3, color="#34A853")
ax2.fill_between(r, r*r*R_2s(r)**2, 0, alpha=0.2, color="#34A853")
ax2.set_xlabel('$r$ [pm]')
ax2.set_ylabel('$r^2 R^2$')
ax2.grid()

base_fsize = 12
plt.rc('axes', labelsize=base_fsize+6)
plt.rc('xtick', labelsize=base_fsize)
plt.rc('ytick', labelsize=base_fsize)
plt.tight_layout(w_pad=2)
plt.show()
```

## 3. Orbital $2p: n=2, l=1, m_l=0$

```python
# componente radial para el orbital 2p
K_2p = 1/(2*np.sqrt(6)) * np.sqrt(a0_inv**3)
def R_2p(r):
    return K_2p * (r/a0) * np.exp(-r/(2*a0))
```

```python
r = np.linspace(0, 800, 100) # picometros

ax1 = plt.subplot(121)
ax1.plot(r, R_2p(r), linewidth=3, color="#20639F")
ax1.set_xlabel('$r$ [pm]')
ax1.set_ylabel('$R$')
ax1.grid()
# ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~
ax2 = plt.subplot(122)
ax2.plot(r, r*r*R_2p(r)**2, linewidth=3, color="#34A853")
ax2.fill_between(r, r*r*R_2p(r)**2, 0, alpha=0.2, color="#34A853")
ax2.set_xlabel('$r$ [pm]')
ax2.set_ylabel('$r^2 R^2$')
ax2.grid()

base_fsize = 12
plt.rc('axes', labelsize=base_fsize+6)
plt.rc('xtick', labelsize=base_fsize)
plt.rc('ytick', labelsize=base_fsize)
plt.tight_layout(w_pad=2)
plt.show()
```
