# Partícula en una caja. Ejercicios

La función de onda que describe a la partícula en una caja es:

$$\displaystyle \Psi_n \left(x\right) = \left( \frac{2}{l} \right)^{1/2} sin\left(\frac{n \pi}{l} x\right)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math, Markdown
%pylab inline

def latexp(str_): display(Markdown(str_))
```

```python
# Funcion de onda para particula en una caja
def Psi_n(x, n, l):
    return np.sqrt(2/l) * np.sin(n*np.pi*x/l)
```

```python
pylab.rcParams['figure.figsize'] = (12.0, 4.0)
```

## 1.1 Gráficas de $\Psi_n \quad \forall n \in \left\{1,2,3\right\}$

```python
l = 1.0
epsilon = 0.01
x = np.arange(0.0, l+epsilon, epsilon)

fig, axs = plt.subplots(1, 3)
color = ["#20639F", "#34A853", "#900C3F"]
for i, n in enumerate([1,2,3]):
    axs[i].axhline(y=0, color="black")
    axs[i].plot(x, Psi_n(n, x, l), linewidth=3, color=color[i])
    axs[i].axis(xmin=0-epsilon,xmax=l+epsilon)
    axs[i].set_xticks(np.linspace(0,l,6))
    axs[i].set_xticklabels([0, "", "", "", "", l])
    axs[i].set_title("$n = {}$".format(n))
    axs[i].grid()
axs[0].set_ylabel('$\Psi_n (x)$')
base_fsize = 12
plt.rc('axes', labelsize=base_fsize+4)
plt.rc('axes', titlesize=base_fsize+4)
plt.rc('xtick', labelsize=base_fsize)
plt.rc('ytick', labelsize=base_fsize)
plt.show()
```

## 1.2 Gráficas de $\left|\Psi_n\right|^2 \quad \forall n \in \left\{1,2,3\right\}$

```python
l = 1.0
epsilon = 0.01
x = np.arange(0.0, l+epsilon, epsilon)

fig, axs = plt.subplots(1, 3)
color = ["#20639F", "#34A853", "#900C3F"]
for i, n in enumerate([1,2,3]):
    axs[i].axhline(y=0, color="black")
    axs[i].plot(x, Psi_n(n, x, l)**2, linewidth=3, color=color[i])
    axs[i].fill_between(x, Psi_n(n, x, l)**2, 0, alpha=0.2, color=color[i])
    axs[i].axis(xmin=0-epsilon,xmax=l+epsilon)
    axs[i].set_xticks(np.linspace(0,l,6))
    axs[i].set_xticklabels([0, "", "", "", "", l])
    axs[i].set_title("$n = {}$".format(n))
    axs[i].grid()
axs[0].set_ylabel('$\Psi_n (x)$')
base_fsize = 12
plt.rc('axes', labelsize=base_fsize+4)
plt.rc('axes', titlesize=base_fsize+4)
plt.rc('xtick', labelsize=base_fsize)
plt.rc('ytick', labelsize=base_fsize)
plt.show()
```

**2.4 Para el butadieno los $4e^- \pi$ ocupan los niveles $n=1$ y $n=2$ (cada nivel ocupado por dos electrones) en su estado fundamental. La longitud del butadieno es $l=5.78 A$. Determine:**

```python
# Constantes
l = 5.78e-10 # m - longitud del butadieno
h = 6.626e-34 # kg·m^2 s^-1 - constante de Planck / 2pi
m_e = 9.10938291e-31 # kg - masa del electron
c = 299792458.0 # m/s - velocidad de la luz
```

```python
def E_n(n, l, h, m):
    return (n**2)*(h**2)/(8*m*(l**2))

# energia de un electron en cada nivel
E1 = E_n(1, l, h, m_e)
E2 = E_n(2, l, h, m_e)
E3 = E_n(3, l, h, m_e)

# energia total en estado fundamental
E_t_ef = 2*E1 + 2*E2
# energia total del primer estado excitado
E_t_1ee = 2*E1+E2+E3
# diferencia de energias en Joules
delta_E = E_t_1ee - E_t_ef
# diferencia de energias en cm^-1
lambda_inv = delta_E/(h*c)/(100)
# error vs frecuencia experimental, absoluto y porcentual
lambda_inv_e = 46100
e_abs = lambda_inv_e-lambda_inv
perror = (lambda_inv_e-lambda_inv)/lambda_inv_e*100
```

**a) La energía asociada a cada nivel.**

La energía de un electrón en el nivel $n$ esta dada por:

$$E_n = \frac{n^2 h^2}{8m l^2}$$

Para el caso del Bitadieno tenemos que la energía de un electrón en cada nivel es:

+ $E_1 =$ {{r'${:0.5E}$'.format(E1)}} $[\text{J}]$
+ $E_2 =$ {{r'${:0.5E}$'.format(E2)}} $[\text{J}]$
+ $E_3 =$ {{r'${:0.5E}$'.format(E3)}} $[\text{J}]$

Así, la energía asociada a cada nivel es:

**Nivel $n=1$**:

$E_{T|n=1} = 2 E_1 =$ {{r'${:0.5E}$'.format(2*E1)}} $[\text{J}]$

**Nivel $n=2$**:

$E_{T|n=2} = 2 E_2 =$ {{r'${:0.5E}$'.format(2*E2)}} $[\text{J}]$

$$------$$

**b) La energía total en el estado fundamental $\left(E_{T_F}\right)$.**

$E_{T_F} = 2E_1 + 2E_2 =$ {{r'${:0.5E}$'.format(E_t_ef)}} $[\text{J}]$

$$------$$

**c) La energía del primer estado exitado $\left(E_{T_{1e}}\right)$ debido a la promoción de $1e^-$ desde $n=2$ a $n=3$**

$E_{T_{1e}} = 2E_1 + E_2 + E_3 =$ {{r'${:0.5E}$'.format(E_t_1ee)}} $[\text{J}]$

$$------$$

**d) La diferencia de energía de los estados $\Delta E$.**

$$------$$

**e) Calcule el error con respecto a la banda  de absorción observada para el butadieno $\left(46,100 \;\text{cm}^{-1}\right)$.**

```python
def E_n(n, l, h, m):
    return (n**2)*(h**2)/(8*m*(l**2))

E1 = E_n(1, l, h, m_e)
E2 = E_n(2, l, h, m_e)
E3 = E_n(3, l, h, m_e)
latexp(r"""La energía de un electrón en cada nivel es:

$E_1 = {:0.5E} \quad [\text{{J}}] \\
E_2 = {:0.5E} \quad [\text{{J}}] \\
E_3 = {:0.5E} \quad [\text{{J}}]$""".format(E1, E2, E3))
```

**a) La energía asociada a cada nivel.**

```python
latexp(r"""La energía asociada a cada nivel es:

$n=1, E_{{T|n=1}} = 2 E_1 = {:0.5E} \quad [\text{{J}}]\\
n=2, E_{{T|n=2}} = 2 E_2 = {:0.5E} \quad [\text{{J}}]$""".format(2*E1, 2*E2))
```

**b) La energía total en el estado fundamental $\left(E_{T_F}\right)$.**

```python
E_t_ef = 2*E1 + 2*E2
latexp(r"$E_{{T_F}} = {:0.5E} \quad [\text{{J}}]$".format(E_t_ef))
```

**c) La energía del primer estado exitado $\left(E_{T_{1e}}\right)$ debido a la promoción de $1e^-$ desde $n=2$ a $n=3$**

```python
E_t_1ee = 2*E1+E2+E3
latexp(r"$E_{{T_{{1e}}}} = {:0.5E} \quad [\text{{J}}]$".format(E_t_1ee))
```

**d) La diferencia de emergía de los estados $\Delta E$.**

```python
delta_E = E_t_1ee - E_t_ef
latexp(r"$\Delta E = {:0.5E} \quad [\text{{J}}]$".format(delta_E))
```

Tranformando $$

```python
lambda_inv = delta_E/(h*c)/(100)
latexp(r"$\Delta E = {:0.2f} \quad [\text{{cm}}^{{-1}}]$".format(lambda_inv))
```

**e) Calcule el error con respecto a la banda  de absorción observada para el butadieno $\left(46,100 \;\text{cm}^{-1}\right)$.**

```python
lambda_inv_e = 46100
perror = (lambda_inv_e-lambda_inv)/lambda_inv_e*100
latexp(r"$\epsilon = {:0.2f} \quad [\text{{cm}}^{{-1}}] \\ \delta = {:0.2f}\%$".format(lambda_inv_e-lambda_inv, perror))
```
