# Gráficas de la componente radial de la solución de la ecuación de Schrödinger para el átomo de Hidrógeno

```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
pylab.rcParams['figure.figsize'] = (10.0, 10.0)
```

```python
# 1s

c_1s = 1/np.sqrt(4*np.pi)

def Y_1s(theta, phi):
    return 0*theta + c_1s
```

```python
theta = np.linspace(0, 2*np.pi, 100)
phi = 0
thetaG, phiG = np.meshgrid(theta, phi)

r = Y_1s(thetaG.flatten(), phiG.flatten())


# TODO
# incrementar el ancho de linea
# cambiar color: naranja, morado
# ajustar tamaño de imagen
# incluir título

# plot
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)
ax.set_rmax(0.5)
ax.set_rticks([])
#ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

plt.show()
```

```python
theta = np.linspace(0, 2*np.pi, 100)
phi = 0
thetaG, phiG = np.meshgrid(theta, phi)
thetaGf, phiGf = thetaG.flatten(), phiG.flatten()

r = Y_1s(thetaG, phiG)

# cartesian coords
x = r * np.sin(thetaGf) * np.cos(phiGf)
y = r * np.sin(thetaGf) * np.sin(phiGf)
z = r * np.cos(thetaGf)
```

```python
plt.plot(x.flatten(), z.flatten())
plt.grid()
plt.show()
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='viridis',edgecolor='none')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
